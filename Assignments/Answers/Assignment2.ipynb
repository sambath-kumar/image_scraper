```python
# Answer 2
# Boolean operators are 
# and
# or 
# not

a1 = True and False
a2 = False or True
a3_ = not(False)

print(a1, a2, a3_)  # a1 perform 'and', a2 perform 'or', a3 perform 'not'
```

```python
# 'and' operator: Produce True result when both inputs are True otherwise Fasle
and1 = False and False
and2 = False and True
and3 = True and False
and4 = True and True
print(and1, and2, and3, and4)

# 'or' operator: Produce True result when any input is True otherwise Fasle
or1 = False or False
or2 = False or True
or3 = True or False
or4 = True or True
print(or1, or2, or3, or4)

# 'not' operator: Inverse of input
not1 = not(True)
not2 = not(False)
print(not1, not2)
```

```python
# Answer 11
# synatx: range(start, end, step)

print("output for range(10):")
# it has end=10, by default start and step values are: 0,1
for i in range(10):
    print(i)
print("output for range(0,10):")
# it has start=0,end=10, by default start and step values are: 0,1
for i in range(0,10):
    print(i)
print("output for range(0,10,1):")
# it has start=0,end=10, by default start and step values are: 0,1
for i in range(0,10,1):
    print(i)
    
# irrespective start and step values, range function produces same results
```
